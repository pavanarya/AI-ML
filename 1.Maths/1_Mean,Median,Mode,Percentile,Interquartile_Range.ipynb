Mean : Average of all items in a dataset

Median : If we have n items in dataset then median = (n+1)th/2 (Note: n=odd)
((n/2)th+((n/2)+1)th)/2 (when n=even)
image.png

25th percentile : ((25*total_items)/100) th element
image.png

75th percentile : ((75*total_items)/100) th element

IQR - Interquartile Range
Outlier : The interquartile range is often used to find outliers in data. Outliers here are defined as observations that fall below Q1  1.5 IQR or above Q3 + 1.5 IQR
(q1 lets say 25th Quartile, Q3 =75th Quartile)
IQR = q3-q1

Missing value fill using Mean, Median

Mode : Most frequently occurring value in a dataset

```python
import pandas as pd
import numpy as np
```

```python
df_airbnb = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/content/AB_NYC_2019.csv')
```

```python
df_airbnb.describe()
```

```python
quantile_25 = df_airbnb.price.quantile(0.25) #25th percentile
quantile_25
```

```python
quantile_35 = df_airbnb.price.quantile(0.35) #35th percentile
quantile_35
```

```python
quantile_75 = df_airbnb.price.quantile(0.75) #75th percentile
quantile_75
```

```python
quantile_80 = df_airbnb.price.quantile(0.80) #80th percentile
quantile_80
```

```python
quantile_99 = df_airbnb.price.quantile(0.99) #99th percentile
quantile_99
```

```python
q_10,q_25,q_30,q_50,q_75,q_99 = df_airbnb.price.quantile([0.10,0.25,0.30,0.50,0.75,0.99])
q_10,q_25,q_30,q_50,q_75,q_99
```

```python
df_airbnb = df_airbnb[(df_airbnb.price>=quantile_25) & (df_airbnb.price<=quantile_80)]
df_airbnb.describe()
```
